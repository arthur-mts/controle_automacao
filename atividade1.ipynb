```python
V = 1
J = 2
b = 0.5
La =1e-3
Ra = .4
Kb = 0.0274
Kt = 0.0274
Km = 10
Td =0
Va =1
```

```python
modelo_eletrico = control.TransferFunction(1, [La, Ra])
modelo_mecanico = control.TransferFunction(1, [J, b])
modelo_velocidade = control.TransferFunction(1, [1, 0])
```

```python
t, y = control.step_response(control.feedback(modelo_eletrico, modelo_mecanico, 1), T_num=10000)
```

```python
plt.plot(t, y)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rad/s)')
plt.title('Resposta ao degrau do motor DC')
plt.show()
```
