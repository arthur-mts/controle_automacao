## Capitulo 4: Caracteristicas de Sistema de Controlecom retroação

A retroação é um elemento crucial nos sistemas de controle, sendo o sinal de erro um componente fundamental nesse processo. O objetivo principal é minimizar esse sinal de erro, buscando alcançar o desempenho desejado. Além disso, é importante mitigar os efeitos de variações indesejadas nos parâmetros do sistema, garantindo sua estabilidade e precisão. Outro aspecto relevante é a redução de sinais de perturbação na saída do sistema, visando obter resultados mais consistentes. É válido ressaltar que o custo associado a um sistema de retroação está relacionado ao dispositivo sensor utilizado nesse processo.


### Sistemas de malha aberta e fechada

Um sistema de controle pode ser classificado como malha aberta ou malha fechada, dependendo da presença ou ausência de retroação.

Em um sistema de malha aberta (ou sistema direto), o controle é realizado sem levar em consideração o feedback ou a resposta real do sistema. Nesse caso, o sinal de controle é determinado apenas com base nas informações de entrada e na função de transferência do sistema. Um exemplo de sistema de malha aberta é uma lâmpada controlada por um interruptor. Ao acionar o interruptor, a lâmpada é ligada, independentemente da iluminação ambiente.

Já um sistema de malha fechada (ou sistema com retroação) leva em conta o feedback do sistema para realizar o controle. É gerado um sinal de erro, que representa a diferença entre a resposta desejada e a resposta real do sistema. Esse sinal de erro é utilizado para ajustar o sinal de controle e alcançar um desempenho mais preciso e estável. Um exemplo comum de sistema de malha fechada é o controle de temperatura de um forno. Sensores medem a temperatura real do forno e comparam com a temperatura desejada. Com base nessa diferença, o sistema de controle ajusta a entrada de energia para manter a temperatura dentro dos parâmetros desejados.

Em resumo, enquanto um sistema de malha aberta não considera o feedback do sistema, um sistema de malha fechada utiliza a retroação para ajustar o controle e alcançar o resultado desejado.

### Sensibilidade de sistemas de controle a variação de parâmetros

Um sistema de controle enfrenta desafios como o envelhecimento do processo, a falta de conhecimento sobre os valores exatos dos parâmetros e outros fatores naturais que podem afetar o procedimento de controle. Em sistemas de malha aberta, esses erros resultam em modificações e imprecisões na saída, tornando a sensibilidade do sistema às variações de parâmetros uma questão fundamental.

Uma maneira útil de lidar com isso é aumentar a magnitude da função de transferência da malha, representada por G(s)H(s). Assim, a principal vantagem de um sistema com retroação é a redução das variações nos parâmetros do processo (G(s)).

A sensibilidade do sistema é definida como a relação entre a mudança na função de transferência e a mudança na função de transferência do processo (ou parâmetro) para uma pequena mudança incremental. Em outras palavras, mede a resposta do sistema a variações nos parâmetros.

Portanto, ao adicionar um loop de controle, é possível reduzir o efeito das variações nos parâmetros de um sistema de controle, o que se torna crucial nos sistemas de controle com retroação.

O conceito de sensibilidade também pode ser encontrado em amplificadores operacionais e está relacionado à capacidade de resposta do amplificador a mudanças nos componentes e parâmetros do circuito.






### Controle da resposta transitória de sistemas

A resposta transitória de um sistema é a resposta do sistema ao longo do tempo. Em um sistema de controle, a resposta transitória é influenciada pelos parâmetros e configuração da malha de controle.

Em um sistema de malha fechada, é possível ajustar os parâmetros da malha de retroação para obter a resposta desejada. Isso permite um maior controle sobre a resposta transitória, buscando um tempo de resposta mais rápido, menor oscilação ou outras características desejadas.

No caso de um sistema de malha aberta, onde não há retroação, é possível modificar a resposta transitória através da inserção de um controlador em cascata adequado. Isso significa adicionar um componente adicional ao sistema para influenciar sua resposta.

Se a resposta transitória de um sistema, seja em malha aberta ou fechada, for muito lenta e não atender aos requisitos desejados, uma opção é escolher outro componente do sistema, como um motor, com uma constante de tempo diferente. Isso implica substituir o elemento que está causando a lentidão por um com uma resposta mais rápida, se possível, a fim de melhorar a resposta transitória do sistema.

Em resumo, a resposta transitória de um sistema pode ser ajustada através dos parâmetros da malha de controle em um sistema de malha fechada ou pela escolha adequada de componentes em um sistema de malha aberta, buscando alcançar uma resposta mais rápida e adequada aos requisitos do sistema.






### Sinais de perturbação em um sistema de controle com retroação

Primeiramente, é importante dizer que: um sinal de perturbação é um sinal de entrada indesejável que afeta o sinal de saída do sistema. É um sinal que acontece em diversos sistemas de controle, produzindo uma saída inexata.

Uma das principais razões de se inserir a retroação é a capacidade de diminuir os efeitos de perturbações e de ruídos que ocorrem dentro da malha de retroação.
É evidente que o esforço para reduzir os efeitos de variações de parâmetros ou de perturbações são proporcionais, e elas se reduzem simultaneamente.

### Erro de estado estacionário

O sistema de controle com retroação é importante, pois oferece ao engenheiro o recurso de ajustar a resposta transitória. Deve-se realizar uma comparação do erro final de estado estacionário de um sistema a malha aberta com o de um sistema a malha fechada.
O erro de estado estacionário é o erro depois que a resposta transitória desapareceu, deixando apenas a resposta contínua.


Para calcular o erro de estado estacionário, se utiliza o teorema do valor final, ou seja:

$$ \lim_{t \rightarrow \infty} e(t) =\lim_{s \rightarrow 0} sE(s)  $$


Um sistema de controle em malha fechada reduz o erro estacionário causado por alterações nos parâmetros ou erros de calibração. Exemplos incluem controle térmico, regulação de tensão e controle de nível de água.

### Custos da retroação

O custo da retroação se manifesta primeiro no aumento do número de componentes e na complexidade do sistema. O sensor é um elemento chave para a retroação. O sensor geralmente é o componente mais caro em um sistema de controle.

O segundo custo da retroação é a perda de ganho. Corriqueiramente, existe ganho a malha aberta com sobra, e se deseja trocar isto por um aumento no controle da resposta do sistema.

Outro custo da retroação é a introdução da possibilidade de instabilidade. Embora o sistema a malha aberta seja estável, o sistema a malha fechada pode não ser sempre estável.

## Circuito RC

```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

Vamos modelar uma simulação do RC

```python
R = 20.0e3
C = 10.0e-6
tau = R*C
Temp_simu = 2.0


def RC(K_p = 3.0):
  numerador = [1/tau]
  denominador = [1.0, 1/tau]
  H_s = ctl.tf(numerador, denominador)
  print(f"\n\n> Função de Transferência em Malha Aberta:\n {H_s}")

  C_s=ctl.tf([K_p],[1.])
  print(f"\n> Função de Transferência do Controlador:\n {C_s}")

  P_s=ctl.tf([1.],[1.])

  G_s=ctl.series(C_s, H_s);
  G1_s=ctl.feedback(G_s, P_s, sign=-1);
  print(f"\n> Função de Transferência em Malha fechada:\n {G1_s}")

  T_mf, yout_mf = ctl.step_response(G1_s, Temp_simu)

  plt.figure(figsize=(15, 7))
  plt.plot(T_mf, yout_mf, linewidth = 1.2)

  Temp_deg = np.linspace(-0.2, Temp_simu, 1000)
  degrau = np.ones_like(Temp_deg)
  degrau[Temp_deg < 0] = 0
  plt.plot(Temp_deg, degrau, color = "r", linestyle = "--", linewidth = 1.2)
  if K_p > 0:
    plt.xticks(np.linspace(0, 2, 20))
    plt.yticks(np.linspace(0, 1, 20))

  plt.title("Circuito RC - Resposta ao Degrau em Malha Fechada", fontsize =  17)
  plt.ylabel("Tensão (V)", fontweight="bold", fontsize = 14)
  plt.xlabel("Tempo (s)", fontweight="bold", fontsize = 14)
  plt.legend(["Resposta ao Degrau", "Degrau Unitário"], fontsize = 14)
  #plt.grid()
  plt.show()

RC(K_p = 3.0)
```

O CRITÉRIO DE ESTABILIDADE DE ROUTH-HURWITZ

A estabilidade de um sistema é determinada pelo critério de Routh-Hurwitz, que analisa o polinômio característico do sistema. Um sistema é considerado instável se todos os coeficientes desse polinômio forem positivos.

O critério de Routh-Hurwitz é um método necessário e suficiente para avaliar a estabilidade de sistemas lineares. Inicialmente, foi desenvolvido utilizando determinantes, mas posteriormente foi apresentada uma forma mais conveniente usando uma tabela organizada. O critério é baseado na ordem dos coeficientes do polinômio e utiliza um algoritmo que calcula esses coeficientes, baseando-se em determinantes de matrizes ou na forma da equação diferencial.

De acordo com a metodologia de Routh-Hurwitz, o número de raízes da equação polinomial com parte real positiva é igual ao número de mudanças de sinal na primeira coluna da tabela de Routh. Portanto, para que um sistema seja estável, não deve haver mudanças de sinal na primeira coluna. Esse requisito é simultaneamente necessário e suficiente para a estabilidade do sistema.

Através de rearranjos matemáticos, são considerados quatro casos diferentes que podem ocorrer nas equações expressas no tópico:

Nenhum elemento da primeira coluna é igual a zero;
Existe um zero na primeira coluna, mas outros elementos da linha que contém esse zero são diferentes de zero;
Existe um zero na primeira coluna e os outros elementos da linha também são iguais a zero;
O mesmo caso que o anterior, mas com raízes repetidas no eixo jw.
Esses casos são analisados para determinar a estabilidade do sistema de acordo com o critério de Routh-Hurwitz.

# Capitulo 7

O desempenho dinâmico de um sistema de controle em malha fechada é descrito pela função de transferência em malha fechada, que é representada por uma razão de polinômios. Essa função, denotada por T(s), relaciona a saída do sistema (Y(s)) com o sinal de referência (R(s)).

A equação característica do sistema, representada por 1 + KG(s) = 0, é fundamental para determinar os modos de resposta do sistema. Nessa equação, K é um parâmetro variável e G(s) é a função de transferência do processo.

As raízes da equação característica, localizadas no plano s, determinam os modos de resposta do sistema. Para que o sistema seja estável, as raízes devem satisfazer a condição em que |KG(s)| = 1. Isso significa que o produto do ganho K e da função de transferência G(s) deve ter uma magnitude igual a 1.

Além disso, a fase do produto KG(s) deve ser igual a -180° (ou 180° mais um múltiplo inteiro de 360°) para garantir a estabilidade do sistema. Essa condição é expressa na forma polar, onde o ângulo de KG(s) deve ser um múltiplo de 180°.

Em resumo, para que um sistema de controle em malha fechada seja estável, é necessário que o produto KG(s) tenha magnitude igual a 1 e uma fase de -180° (ou 180° mais um múltiplo inteiro de 360°). Essas condições garantem a localização adequada das raízes características do sistema no plano s.
